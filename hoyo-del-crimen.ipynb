```python
import certifi
import urllib3
import json
from pandas.io.json import json_normalize
```

```python
http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',ca_certs=certifi.where())
r = http.request('GET', 'http://hoyodecrimen.com/api/v1/cuadrantes/all/crimes/all/period')
r.status
response = r.data
```

```python
str_response = response.decode('utf-8')
y = json.loads(str_response)
y = y['rows']
```

```python
json_normalize(y)
```
